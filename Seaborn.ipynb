```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.drop(['alone','alive','who','adult_male','embark_town','class'], axis = 1, inplace = True)
titanic.dropna(axis = 0, inplace = True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

Autant utiliser les fonctions represantants au mmieux les variables categorielles car pairplot ne represente pas bien nos variables

```python
sns.catplot(x = 'pclass', y = 'age', data = (titanic), hue = 'sex')
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = (titanic), hue = 'sex')

#Les hommes voyageant en 1 ere classe ont un age median de 40 ans, le maximum est de 50ans et le minimum de 30 ans
```

```python
#Distribution de la variable 'Tarif'

sns.distplot(titanic['fare'])
```

```python
sns.jointplot('age', 'fare', data = titanic, kind = 'hex') #On affiche la distribution selon deux variables
```

```python
# Pour visualiser les correlations :

sns.heatmap(titanic.corr())
```
